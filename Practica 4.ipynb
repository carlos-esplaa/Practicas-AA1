## Ejercicio 2:  Clasificación de muestras.

```python
x,y = datasets.make_blobs(n_samples=100, n_features=2, centers=[[0,0], [2,2], [2,0]], cluster_std=0.6, random_state=123)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, train_size=0.7)
clf = Perceptron().fit(x_train, y_train)
decision = clf.decision_function(x_test)
print(decision)
```

- Con esta muestra aleatoria de casos, cogemos el 70% de la muestra de entrenamiento y el 30% restante de test. Entrenamos un clasificador lineal con la clase Perceptron y obtenemos la clasificación de las muestras de test, utilizando la función decision_funcion.

```python
predict = clf.predict(x_test)
print(predict, '\n', y_test)
correct = 0
for i in range(len(predict)):
    if predict[i] == y_test[i]:
        correct += 1
(correct/len(y_test))*100
```
